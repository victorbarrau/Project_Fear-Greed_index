```python
import pandas as pd
import numpy as np

# Générer 100 valeurs aléatoires entre 0 et 100 pour la colonne "trends"
trends = np.random.randint(0, 101, size=100)

# Créer un dataframe avec la colonne "trends"
df = pd.DataFrame({'trends': trends})
```

```python
# Calculer la moyenne mobile de 15 valeurs
rolling_mean = df['trends'].rolling(window=14).mean()

# Ajouter la colonne "moyenne mobile" au dataframe
df['moving_avg'] = rolling_mean
```

```python
# Calculer l'écart à la moyenne mobile
df['moving_avg_diff'] = df['trends'] - rolling_mean
```

```python
# Normaliser la colonne "moving_avg_diff"
df['moving_avg_diff_normalized'] = (df['moving_avg_diff'] + 100) / 200
```

```python
import matplotlib.pyplot as plt



# Tracer la colonne "écart à la moyenne mobile" normalisée
plt.plot(df['moving_avg_diff_normalized'], label='Moving Average Diff (Normalized)')

# Ajouter une légende et un titre
plt.legend()
plt.title('Trends and Moving Averages')

# Afficher le graphique
plt.show()
```

```python
import matplotlib.pyplot as plt

# Tracer la colonne "trends"
plt.plot(df['trends'], label='Trends')

# Tracer la colonne "moyenne mobile"
plt.plot(df['moving_avg'], label='Moving Average')

# Ajouter une légende et un titre
plt.legend()
plt.title('Trends and Moving Averages')

# Afficher le graphique
plt.show()
```
